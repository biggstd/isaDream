```python
class QueryGroup(param.Parameterized):
    
    label = param.String(doc="The name of the query group.")
    unit = param.String(doc="The unit of factors to query.")
    species = param.List(doc="Species to match")
    
```

```python
QueryGroup(name='27 Al ppm', 
           unit='ppm', species=["Al"])

QueryGroup(name='Total Aluminate Concentration', 
           unit='Molar', species=["Al"])

QueryGroup(name='Counter Ion Concentration', 
           unit='Molar', species=["Na+", "Li+", "Cs+", "K+"])

QueryGroup(name='Counter Ion', 
           unit='Species', species=["Na+", "Li+", "Cs+", "K+"])

QueryGroup(name='Base Concentration', 
           unit='Molar', species=["OH-"])
```

```python
widgets = parambokeh.Widgets(QueryGroup)
```

```python
class FactorElement(param.Parameterized):
    
    factor_type = param.String(
        default="Measurement Condition",
        doc=textwrap.dedent("""\
        Factor categorization. A `Measurement Condition` indicates 
        a value an experimentalist controls or sets.
        A `Measurement` is a value obtained from a given experiment
        or simulation.
        """))
    
    reference_value   = param.String(default=None,
                                     doc="A value which references a controlled term.")
    
    string_value      = param.String(default=None,
                                     doc="A value which can be any string. For uncontrolled terms.")
    
    decimal_value     = param.Number(default=None,
                                     doc="A value which can be any number.")
    
    unit_reference    = param.String(default=None,
                                     doc="A value which references a controlled unit term.")
    
    csv_column_index  = param.Integer(default=None,
                                      doc="A value which references the column index of the current assay's data file which this factor labels.")
```
