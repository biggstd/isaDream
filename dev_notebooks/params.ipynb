# Paramaterized Elemental Classes

```python
import param
```

```python
class Factor(param.Parameterized):
    
    allowed_factor_types = ["Measurement Condition", "Measurement",
                            "Measurement Reference"]
    
    factor_type = param.ObjectSelector(default="Measurement Condition",
                                       objects=allowed_factor_types,
                                       doc="Factor type documentation.")
    decimal_value = param.Number(default=None)
    string_value  = param.String()
    
    reference_value = param.String()
    unit = param.String()
    csv_column_index = param.Integer()
```

```python
f = Factor()
```

```python
import parambokeh
from bokeh.io import output_notebook
output_notebook()
```

```python
widgets = parambokeh.Widgets(Factor)
```

```python
class SpeciesFactor(param.Parameterized):
    species_references = param.String()
    stoichiometry = param.Number(default=1.0)
```

```python
widgets = parambokeh.Widgets(SpeciesFactor)
```

```python
class FactorCollection(param.Parameterized):
    factors    = param.List()
    add_factor = param.Action(lambda x: x.factors.append())
```
